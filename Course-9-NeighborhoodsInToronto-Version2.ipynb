```python
import pandas as pd

#Read data html table into data frames
dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

#Get the first table from the list of tables
df = dfs[0]
df
```

```python
df.describe()
```

```python
#Filter to select rows that have Borough value != 'Not assigned'
filterDf = df[df.Borough != 'Not assigned']
filterDf
```

```python
aggregateFunc = {'Postcode':'first', 'Borough':'first','Neighbourhood':lambda x: "%s " % ','.join(x)}
#groupDf = filterDf.groupby(['Postcode', 'Borough']).aggregate(aggregateFunc).reindex(columns=['Postcode', 'Borough', 'Neighbourhood'])
groupDf = filterDf.groupby(['Postcode', 'Borough']).aggregate(aggregateFunc).reindex(columns=['Postcode', 'Borough','Neighbourhood'])
groupDf['Latitude'] = ''
groupDf['Longitude'] = ''
groupDf
```

```python
groupDf.shape
```

```python
coordDf = pd.read_csv("http://cocl.us/Geospatial_data")
coordDf.columns = ['Postcode', 'Latitude', 'Longitude']
coordDf.reindex(['Postcode'])
coordDf.head()
```

```python
# Function to search for coords based on Postcode
def searchCoords(Postcode):
    for idx, row in coordDf.iterrows():
        if(row['Postcode'] == Postcode):
            return row['Latitude'], row['Longitude']
```

```python
for idx, row in groupDf.iterrows():
    lat, lng = searchCoords(row['Postcode'])
    row['Latitude'] = lat
    row['Longitude'] = lng
    # print(lat,lng)

groupDf
```

```python
address = 'Toronto, Ontario'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python

# Create map of toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(groupDf['Latitude'], groupDf['Longitude'], groupDf['Borough'], groupDf['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto

```
